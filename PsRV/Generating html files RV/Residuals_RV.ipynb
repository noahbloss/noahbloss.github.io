```python
import os
from IPython.display import display, HTML


# Path to the directory containing the plots
plot_dir = "C:\\Users\\noahb\\Desktop\\GithubPagesFiles\\PsRV\\CorrelationRV\\Plots_Relations_within_the_SeasonRV"

# List all plot files in the directory
plot_files = [f for f in os.listdir(plot_dir) if f.startswith('Plot_') and f.endswith('.png')]

# Extract numeric parts and sort the plot files numerically
plot_files.sort(key=lambda f: int(f.split('_')[1].split('.')[0]))

# Generate HTML content dynamically
html_content = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Relations within the Season</title>
    <style>
        body {
            font-family: Arial, sans-serif;
        }
        .image-container {
            width: 60%;
            margin: 0 auto;
            text-align: center;
            margin-bottom: 35px;
        }
        .image-container img {
            width: 100%;
            max-width: 100%;
        }
        .home-button {
            position: fixed;
            top: 10px;
            right: 10px;
            background-color: #4CAF50;
            color: white;
            padding: 10px 20px;
            text-decoration: none;
            border-radius: 5px;
        }
        .home-button:hover {
            background-color: #45a049;
        }
    </style>
</head>
<body>
    <a href="/index.html" class="home-button">Main Page</a>
    <h1>Relations within the Season</h1>
    <h5> This Page shows the Correlation between the different Sabermetrics within the Season over the Years. 
    To have a better oversight over the Correlation Coefficients, the Table with all Coefficients is provided in 
    <a href="Correlation_Table_RV.html">Correlation Tables</a>. 
    The first Plots, before the horizontal rule, are created with Data drawn from pitchers with at least 300 pitches. 
    After that the same Plots are displayed, except the Plots were created with Data drawn from Pitchers with at least 1750 pitches. 
    </h5>
"""

# Add each plot to the HTML content
for plot_file in plot_files:
    plot_path = os.path.join('..', 'CorrelationRV', 'Plots_Relations_within_the_SeasonRV', plot_file)  # Relative path to plot
    html_content += f'<div class="image-container"><img src="{plot_path}" alt="{plot_file}"></div>'
    html_content += '<hr>'  # Horizontal line between plots for separation

html_content += """
</body>
</html>
"""

# Path to save the final HTML file
output_dir = "C:\\Users\\noahb\\Desktop\\GithubPagesFiles\\PsRV\\htmlRV"
html_file_path = os.path.join(output_dir, 'Relations_within_the_Season_RV.html')

# Save the final HTML content to a file
with open(html_file_path, 'w') as html_file:
    html_file.write(html_content)

# Display the HTML content
display(HTML(html_content))

print(f"HTML file created: {html_file_path}")
```
