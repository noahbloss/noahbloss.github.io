```python
import os
from jinja2 import Template

# Path to the directory containing the HTML table
table_dir = "C:\\Users\\noahb\\Desktop\\GithubPagesFiles\\DataTables"
table_file = os.path.join(table_dir, 'pitchers_table.html')

# Read the HTML table
with open(table_file, 'r') as file:
    table_html = file.read()

# HTML template for the page
html_template = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Pitchers Data</title>
    <style>
        body {
            font-family: Arial, sans-serif;
        }
        .home-button {
            position: fixed;
            top: 10px;
            right: 10px;
            background-color: #4CAF50;
            color: white;
            padding: 10px 20px;
            text-decoration: none;
            border-radius: 5px;
        }
        .home-button:hover {
            background-color: #45a049;
        }
    </style>
</head>
<body>
    <a href="/index.html" class="home-button">Overview</a>
    <h1>Pitchers Data</h1>
    {{ table_html | safe }}
</body>
</html>
"""

# Create the final HTML by rendering the template with the table HTML
template = Template(html_template)
html_content = template.render(table_html=table_html)

# Path to save the final HTML file
output_dir = "C:\\Users\\noahb\\Desktop\\GithubPagesFiles\\DataPages"
html_file_path = os.path.join(output_dir, 'pitchers_data.html')

# Save the final HTML content to a file
with open(html_file_path, 'w') as file:
    file.write(html_content)

print(f"HTML file created: {html_file_path}")
```
