```python
import numpy as np
import os
from IPython.display import display, HTML

# Path to the directory containing the plots
plot_dir = "C:\\Users\\noahb\\Desktop\\GithubPagesFiles\\Correlation\\Plots_one_relation_over_the_years"

# List all plot files in the directory
plot_files = [f for f in os.listdir(plot_dir) if f.startswith('Plot_') and f.endswith('.png')]

# Extract numeric parts and sort the plot files numerically
plot_files.sort(key=lambda f: int(f.split('_')[1].split('.')[0]))

# Generate HTML content dynamically
html_content = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>One Correlation over the Years</title>
    <style>
        body {
            font-family: Arial, sans-serif;
        }
        .image-container {
            width: 48%;
            display: inline-block;
            margin-right: 2%;
            margin-bottom: 35px;
            vertical-align: top;
            text-align: center;
        }
        .image-container img {
            width: 100%;
        }
        .clearfix::after {
            content: "";
            clear: both;
            display: table;
        }
        .home-button {
            position: fixed;
            top: 10px;
            right: 10px;
            background-color: #4CAF50;
            color: white;
            padding: 10px 20px;
            text-decoration: none;
            border-radius: 5px;
        }
        .home-button:hover {
            background-color: #45a049;
        }
    </style>
</head>
<body>
    <a href="/index.html" class="home-button">Main Page</a>
    <h1>Correlation Plots</h1>
    <h2>This Page shows the Correlation between one pair of sabermetrics over the years. One Pair is represented in each (9x9)-Plot. </h2>
    <div class="clearfix">
    <ul>
"""

html_content += """
    <p> The exact same Plots are shown in 
    <a href="First_Thing.html">Relations of metrics the following year</a> 
    just ordered differently. One (9x9)-Plot represents the correlation between the two sabermetrics in focus. 
    The sabermetric of one year is only compared to the other sabermetric in focus in the following year. 
    On the Left side the Correlation is calculated from Data drawn from pitchers with at least 300 Pitches. 
    On the right side with at least 1750 pitches. To have a better oversight over the Correlation Coefficients, 
    the Table with all Coefficients is provided in 
    <a href="Correlation_Table.html">Correlation Tables</a>.
    </p>
    """

# Add each plot to the HTML content
for i, plot_file in enumerate(plot_files):
    plot_path = os.path.join('Plots_one_relation_over_the_years', plot_file)  # Relative path to plot
    html_content += f'<div class="image-container"><img src="{plot_path}" alt="{plot_file}"></div>'

    # Add a clearfix div every two images to ensure correct layout
    if (i + 1) % 2 == 0:
        html_content += '<div class="clearfix"></div>'

html_content += """
    </div>
</body>
</html>
"""




output_dir = os.path.join(os.path.expanduser('~'), 'Desktop\GithubPagesFiles\Correlation')
print(output_dir)
html_file_path = os.path.join(output_dir, 'one_relation_over_the_years.html')
print(html_file_path)
with open(html_file_path, 'w') as html_file:
    html_file.write(html_content)

display(HTML(html_content))


```
